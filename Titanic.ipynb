```python
def impute_age(cols):
    pc = cols[0]
    a = cols[1]
    
    if pd.isnull(a):
        if pc == 1:
            return 38
        elif pc == 2:
            return 30
        else:
            return 25
    else:
        return a
```

```python
data['Age'] = data[['Pclass', 'Age']].apply(impute_age,axis=1)
```
